```python
#Changing Data type of Gross and Budget
```

```python
df['budget'] = df['budget'].fillna(0)
```

```python
df['gross'] = df['gross'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Sorting
```

```python
df = df.sort_values(by = ['gross'], inplace= False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any Duplicates
```

```python
df = df['company'].drop_duplicates.sort_values(ascending=False)
```

```python
#Company : numerizing it
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix1 = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix1, annot = True)

plt.title('Correlation Matrix for ALL Features')
```

```python
df_numerized.corr()
```

```python
corr_matrix = df_numerized.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
sorted_pairs.dtype
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
